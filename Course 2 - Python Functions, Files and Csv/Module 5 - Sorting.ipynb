## Sorting Basics

### Invoking.sort and sorted

```python
# Example: .sort()

L1 = [1, 7, 4, -2, 3]
L2 = ['Cherry', 'Apple', 'Blueberry']

# Sorting L1
L1.sort()
print(L1)

# Sorting L2
L2.sort()
print(L2)
```

```python
# Example2:  Sorted()
L1 = [1, 7, 4, -2, 3]
L2 = ['Cherry', 'Apple', 'Blueberry']

print(L2)
print(sorted(L2))
print(L2)  # Unchanged

print('============')


L2.sort()
print(L2)
print(L2. sort())  # return Value is None
```

### Optional Reverse Parameter

```python
# Example 1: Reverse Paremeter
L2 = ['Cherry', 'Apple', 'Blueberry']
print(sorted(L2, reverse=True))


# Example 2: Reverse Paremeter
L1 = [1, 7, 4, -2, 3]
print(sorted(L1, reverse=True))
```

### Optional Key Parameter

```python
# Example 1: Optional Key Parameter
L1 = [1, 7, 4, -2, 3]

def absolute(x):
    if x >= 0:
        return x
    else:
        return -x

print(absolute(3))
print(absolute(-119))

for y in L1:
    print(absolute(y))
```

```python
L2 = sorted(L1, key= absolute)
print(L2)


# Or in reverse order
print(sorted(L2, reverse= True, key=absolute))
```

```python
# Example 2: 
def square(x):
    return x * x

nums = [5, -2, 3]

print(sorted(nums, key=square))
```

```python
# Example 3: 
names = ["Ali", "Mohamed", "Abdi", "A"]
print(sorted(names, key=len))  # Kula sareysii length-kooda 
```

## Sorting Dictionaries, Breaking Ties

### Sorting Dictionary's keys

```python
# Example:
L = ['E', 'F', 'B', 'A', 'D', 'I', 'I', 'C', 'B', 'A', 'D', 'D']

d = {}

for x in L:
    if x in d:
        d[x] = d[x] + 1
    else:
        d[x] = 1

for x in d.keys():
    print(f'{x} appears {d[x]} times')
```

```python
# Sample data of list
L = ['E', 'F', 'B', 'A', 'D', 'I', 'I', 'C', 'B', 'A', 'D', 'D']

# Empty dictionary
d = {}

for x in L:
    if x in d:
        d[x] = d[x] + 1
    else:
        d[x] = 1

# Sorted
for x in sorted(d, key=lambda k: d[k], reverse=True):
    print(f'{x} appears {d[x]} times')
```

### Breaking Ties

```python
# Example
tups = [('A', 3, 2),
        ('C', 1, 4), 
        ('D', 3, 1), 
        ('A', 2, 4),
        ('C', 1, 2)]

for tup in sorted(tups):
    print(tup)
```

```python
# Example 2:
fruits = ['Peach', 'Kiwi', 'Apple', 'Blueberry', 'Papay', 'Mango']

new_order = sorted(fruits, key= lambda fruit_name: (len(fruit_name), fruit_name)) # From lowest to highest

for fruit in new_order:
    print(fruit)
```

```python
# Example 2:
fruits = ['Peach', 'Kiwi', 'Apple', 'Blueberry', 'Papay', 'Mango']

new_order = sorted(fruits, key= lambda fruit_name: (-len(fruit_name), fruit_name)) # from highest to lowest

for fruit in new_order:
    print(fruit)
```

```python
# 6. Create a function called last_four that takes in a single ID number and returns the last four digits. For example, the number 17573005 should return 3005. Then, use the resulting function to sort the list of ids stored in the variable, ids, from lowest to highest. Save this sorted list in the variable, sorted_ids. Hint: Remember that only strings can be indexed, so conversions may be needed.

def last_four(x):
    return int(str(x)[-4:])

ids = [17573005, 17572342, 17579000, 17570002, 17572345, 17579329]

sorted_ids = sorted(ids, key=last_four)

print(sorted_ids)
```
