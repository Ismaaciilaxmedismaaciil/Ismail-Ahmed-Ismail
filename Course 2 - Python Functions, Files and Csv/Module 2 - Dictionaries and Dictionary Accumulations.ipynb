## Module 2 - Dictionary and Dictionary Accumulations

### Dictionary 
 - Is an ordered collection of Items.
 - There is no first item or second item.
 - They contains a bag of key value pairs.
 - Dictionary stores a key value pairs 

```python
# Creating an empty dictionary
empty_dict = {}
empty_dict['One'] = 'Unos'
empty_dict['Two'] = 'Dos'   
empty_dict['Three'] = 'Tres'
empty_dict['Four'] = 'Cuatro'

print(empty_dict)
```

```python
# Also you can write
empty_dict = {'One': 'Unos', 'Two': 'Dos', 'Three': 'Tres', 'Four': 'Cuatro'}
print(empty_dict)
```

```python
value = empty_dict['One']
print(value)
```

### Dictionary Operations

```python
Inventory = {'Apples': 30, 'Bananas': 45, 'Mangoes': 54}

# Deleting Apples Item
del Inventory['Apples']
print(Inventory)

# Updating  value of Bananas Item
Inventory['Bananas'] = 0
print(Inventory) 

# Checking the length of dictionary
numItems = len(Inventory)
print(numItems)

```

### Dictionary Methods

```python
Inventory = {'Apples': 30, 'Bananas': 45, 'Mangoes': 54}

# Accessing One One Item
print(Inventory['Apples'])

# Iterating Only Keys
for key in Inventory.keys():
    print(key)


# List of keys
print(list(Inventory.keys()))

#List of Items
print(list(Inventory.items()))

# Checking a Banana item inside the dictionary
print('Bananas' in Inventory)

# Checking a Banana item inside the dictionary if statement
if 'Bananas' in Inventory:
    print(Inventory['Bananas'])
else:
    print('We have no Banana')


# Accessing Items using get method
print(Inventory.get('Apples'))
print(Inventory.get('Mangoes'))
```

### Aliasing and Copying with Dictionaries

```python
opposites = {'Up': 'Down', 'Right': 'Wrong', 'True': 'False'}
alias = opposites

print(alias is opposites)

alias['right'] = 'left'

print(opposites['right'])
```

## Dictionary Accumulation

```python
file = open('olympics.csv', 'r')
text = file.read()

t_count = 0
s_count = 0
for char in text:
    if char == 't':
        t_count += 1

    elif char == 's':
        s_count += 1

print(f't: {t_count} Occurencies')
print(f's: {s_count} Occurencies')
```

```python
# Dicitonary Accumulation
file = open('olympics.csv', 'r')

text = file.read()

x = {}

# Initialize counter with 0
x['t'] = 0
x['s'] = 0

# Iteration the text variable
for char in text:
    if char == 't':
        x['t'] += 1  
    
    elif char == 's':
        x['s'] += 1

print(f't: {x['t']} Occurencies')
print(f's: {x['s']} Occurencies')
```

```python
# Another way example
file = open('olympics.csv', 'r')
txt = file.read()

x = {}

for char in txt:
    if char not in x:
        x[char] = 0
    
    x[char] += 1

print(f'{x['t']} Occurancies')
print(f'{x['s']} Occurancies')
print(f'{x['v']} Occurancies')
```

```python
# Opening the file and reading the file
file = open('log.csv', 'r')
txt = file.read()

x = {}

for char in txt:
    if char not in x:
        x[char] = 0
    
    x[char] += 1

print(f't: {x['t']} Occurencies')
print(f's: {x['s']} Occurencies')
print(f'b: {x['b']} Occurencies')
```

```python
# Example 3:
file = open('log.csv', 'r')
txt = file.read()

x = {}

for char in txt:
    if char not in x:
        x[char] = 0
    
    x[char] += 1

for y in x:
    print(f"{x[y]} {y}'s Occurencies")
```

## Csv Module with Dictionaries


### Reading `.csv` Files

#### 1- csv.reader

```python
# Example 1: Csv.reader
import csv

with open('olympics.csv', 'r') as file:
    reader = csv.reader(file)

    for row in reader:
        if row[5] != 'NA':
            print(f'{row[0]}: {row[4]}; {row[5]}')
```

#### 2 - Csv.DictReader
- Automatically reads a first row as a header.
- Creates a dictionaries for each other row, with headers as key.

```python
# Example 2: Csv.DictReader

import csv

with open('olympics.csv', 'r') as file:
    reader = csv.DictReader(file)

    print(reader.fieldnames)
    # Shows first two rows
    print(next(reader))
    print(next(reader))
```

```python
with open('olympics.csv', 'r') as file:
    reader = csv.DictReader(file)

    for row in reader:
        if row['Medal'] != 'NA':
            print(f'{row['Name']}: {row['Event']}; {row['Medal']}')
```

### Writing `.csv` files

#### 1. With csv.writer

```python
# Example 1: with csv writer

olympians = [('Ismail', 22, 'Football'),
             ('Cali', 23, 'BasketBall'),
             ('Samiir', 45, 'Cycling'),
             ('Fatima', 43, 'Swimming'),
             ('Salma', 45, 'Running')]

file = open('New_Olympians.csv', 'w')

writer = csv.writer(file)

writer.writerow(['Name', 'Age', 'Football'])

for olympian in olympians:
    writer.writerow(olympian)

file.close()
```

#### 2. With csv.DictReader
If we have a list of dictionaries rather than list of list or tuples, we can use DictWriter



```python
# Example 2: Csv.DictReader

Dahab_Employees = [
    {'Name': 'Ismail Ahmed Ismail', 'Age': 22, 'Title': 'Data Scientist'},
    {'Name': 'Sakeriye Ahmed Ismail', 'Age': 23, 'Title': 'Human Resource'},
    {'Name': 'Abdullahi Mahomed Said', 'Age': 25, 'Title': 'Accountant'},
    {'Name': 'Fowsiyia Abdisammad Ahmed', 'Age': 23, 'Title': 'Investment manager'},
    {'Name': 'Haliima Hiirad Nuur', 'Age': 30, 'Title': 'Solar System'},
    {'Name': 'Abdinasir Mohamed Farah', 'Age': 45, 'Title': 'Logist Officer'},
    {'Name': 'Abdisammad Abdulqadir Jamma', 'Age': 50, 'Title': 'Security'}
    ]


file = open('New_Dahab_Employees.csv', 'w')

writer = csv.DictWriter(file, Dahab_Employees[0].keys())

writer.writeheader()

for employee in Dahab_Employees:
    writer.writerow(employee)

file.close()
```
