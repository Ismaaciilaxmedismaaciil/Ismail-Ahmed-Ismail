### The while statement

```python
def sumTo(aBound):
    '''Returns the sum of 1+2+3.....n'''

    theSum = 0
    aNumber = 1

    while aNumber <= aBound:
        theSum = theSum+ aNumber
        aNumber = aNumber + 1

    return theSum

print(sumTo(4)) # Output:- 1+2+3+4 = 10
print(sumTo(10))
```

```python
def multiTo(Bound):
    '''Returns the multiplication of 1*2*3*...n'''

    themulti = 1
    number = 1

    while number <= Bound:
        themulti = themulti * number
        number = number + 1

    return themulti


print(multiTo(10))
```

```python
def sumTo(abound):
    '''Retruns the sum of the 1+2+3+...n'''

    thesum = 0 
    
    for number in range(1, abound + 1):
        thesum = thesum + number
    
    return thesum

print(sumTo(4))
print(sumTo(1000))
```
