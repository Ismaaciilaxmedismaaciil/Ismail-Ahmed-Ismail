### The while statement

```python
def sumTo(aBound):
    '''Returns the sum of 1+2+3.....n'''

    theSum = 0
    aNumber = 1

    while aNumber <= aBound:
        theSum = theSum+ aNumber
        aNumber = aNumber + 1

    return theSum

print(sumTo(4)) # Output:- 1+2+3+4 = 10
print(sumTo(10))
```

```python
def multiTo(Bound):
    '''Returns the multiplication of 1*2*3*...n'''

    themulti = 1
    number = 1

    while number <= Bound:
        themulti = themulti * number
        number = number + 1

    return themulti


print(multiTo(10))
```

```python
def sumTo(abound):
    '''Retruns the sum of the 1+2+3+...n'''

    thesum = 0 
    
    for number in range(1, abound + 1):
        thesum = thesum + number
    
    return thesum

print(sumTo(4))
print(sumTo(1000))
```

```python
count = 0

while count <= 15:
    count += 1 

print(count)
```

```python
count = 0
even_nums = []

while count <= 15:
    if count %2==0:
        even_nums.append(count)
    
    count = count + 1

print(even_nums)
```

### The Listener Loop

```python
# Example
theSum = 0
x = -1

while (x != 0):
    x = int(input('Enter number to add up (Enter 0 if no more numbers)'))
    theSum = theSum + x

print(theSum)
```

```python
# Example

def get_yes_or_no(message):
    valid_input = False
    while not valid_input:
        answer = input(message)
        answer = answer.upper()  #Convert to upper case
        if answer == 'Y' or answer == 'N':
            valid_input = True
        else:
            print('Please enter Y for yes or N for no.')

    return answer


response = get_yes_or_no('Do you like lima beans? Y)es or N)o: ')
 
if response == 'Y':
    print('Great! They are very healthy.')
else:
    print('Too bad. If cooked right, they are quite tasty.')
    
```

### Break and Continue

```python
# Break
while True:
    print('This phrase will always print')
    break
    print('Does this phrase print?')

print('We are done with the while loop.')
```

```python
# Another Example
x = 0

while x < 10:
    print('We are incrementing x')
    if x%2==0:
        x += 3
        continue

    if x%3==0:
        x += 5
    
    x += 1

print(f'Done with our loop! x has value: {x}')
```
