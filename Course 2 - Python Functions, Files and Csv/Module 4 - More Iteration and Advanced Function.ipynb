### The while statement

```python
def sumTo(aBound):
    '''Returns the sum of 1+2+3.....n'''

    theSum = 0
    aNumber = 1

    while aNumber <= aBound:
        theSum = theSum+ aNumber
        aNumber = aNumber + 1

    return theSum

print(sumTo(4)) # Output:- 1+2+3+4 = 10
print(sumTo(10))
```

```python
def multiTo(Bound):
    '''Returns the multiplication of 1*2*3*...n'''

    themulti = 1
    number = 1

    while number <= Bound:
        themulti = themulti * number
        number = number + 1

    return themulti


print(multiTo(10))
```

```python
def sumTo(abound):
    '''Retruns the sum of the 1+2+3+...n'''

    thesum = 0 
    
    for number in range(1, abound + 1):
        thesum = thesum + number
    
    return thesum

print(sumTo(4))
print(sumTo(1000))
```

```python
count = 0
even_nums = []

while count <= 15:
    if count %2==0:
        even_nums.append(count)
    
    count = count + 1

print(even_nums)
```

### The Listener Loop

```python
# Example
theSum = 0
x = -1

while (x != 0):
    x = int(input('Enter number to add up (Enter 0 if no more numbers)'))
    theSum = theSum + x

print(theSum)
```

```python
# Example

def get_yes_or_no(message):
    valid_input = False
    while not valid_input:
        answer = input(message)
        answer = answer.upper()  #Convert to upper case
        if answer == 'Y' or answer == 'N':
            valid_input = True
        else:
            print('Please enter Y for yes or N for no.')

    return answer


response = get_yes_or_no('Do you like lima beans? Y)es or N)o: ')
 
if response == 'Y':
    print('Great! They are very healthy.')
else:
    print('Too bad. If cooked right, they are quite tasty.')
    
```

## Advanced Function

### Optional Parameter

```python
#  Example 
def f(x, y, z):
    print(f'x, y, z, are: {x}, {y}, {z}')

f(2, 5, 8)

# f(2, 5) #Error 
```

```python
# Example 2:
initial = 7

def f(x, y=3, z=initial):
    print(f'x, y, z, are: {x}, {y}, {z}')

f(2)
f(2, 5)
f(2, 5, 8)
```

### Mutation And Optional Parameters

```python
# Example:
def f(a, l=[]):
    l.append(a)
    return l

print(f(1))
print(f(2))
print(f(3))
print(f(4, ['Hello']))
print(f(5, ['Hello']))
```

### Keyword Parameter

```python
# Example:
initial = 7
def f(x, y=3, z=initial):
    print(f'x, y, z, are: {x}, {y}, {z}')

f(2)
f(2, 5)
f(2, 5, 8)
f(2, z=8)
```

### Testing Optional Parameter

```python
# Example:
def count_long_words(words, min_length= 5):
    ct = 0
    for word in words:
        if len(word) >= min_length:
            ct += 1
    return ct

test_words = [' ', '1', '12', '123', '1234', '12345']
# assert count_long_words(test_words) == 2
# assert count_long_words(test_words, min_length) == 2
# assert count_long_words(test_words, min_length) == 7
# assert count_long_words(test_words, min_length) == 3
# assert count_long_words(test_words, min_length) == 0
```

```python
import math

def distance(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    return (dx * dx + dy * dy) ** 0.5

assert distance(1,2, 1,2) == 0
assert distance(1,2, 4,2) == 3
assert distance(1,2, 4,6) == 5
assert distance(0,0, 1,1) == 2 ** 0.5
```
