## Csv Output

### Reading a csv File

```python
# Readign a csv file
file = open('olympics.csv', 'r')

lines = file.readlines()

for line in lines[:6]:
    print(line.strip())

print('============')

header = lines[0]

fieldnames = header.strip().split(',')
print(fieldnames)

for row in lines[1:]:
    vals = row.strip().split(',')

    if vals[5] != 'NA':
        print(f'{vals[0]}: {vals[4]}; {vals[5]}')
```

### Reading a csv file Using csv module

```python
import csv

file = open('olympics.csv', 'r')
reader = csv.reader(file)

rows = list(reader)

# header row
header = rows[0]
print(header)

for row in rows[1:]:
    if row[5] != 'NA':
        print(f'{row[0]}: {row[4]}; {row[5]}')
```

#### More Elegant Version

```python
import csv

with open('olympics.csv', 'r') as file:
    reader = csv.reader(file)

    # header row
    header = next(reader)
    print(header)

    for row in reader:
        if row[5] != 'NA':
            print(f'{row[0]}: {row[4]}; {row[5]}')
```

```python
file = open('log.csv', 'r')

lines = file.readlines()

header = lines[0]
print(header)

for line in lines[1:]:
    print(line.strip())
```

```python
import csv

with open('log.csv', 'r') as file:
    reader = csv.reader(file)

    # header row
    header = next(reader)
    print(header)

    # each row
    for row in reader:
        print(row)
```

### Writing to files with the csv module

#### By Hand

```python
olympians = [('ISMAIL AHMED ISMAIL', 22, 'FOOTBALL'),
             ('MOHAMED ABDI JAMMA', 32, 'BASKETBALL'),
             ('CALI FARAX MOHAMED', 30, 'CYCLING'),
             ('BASHIIR CALI CABDI', 45, 'RUNNING'),
             ('FARXIA SAMATAR BADI', 23, 'SWIMMING')]


# opening the file
file = open('reduced_olympians.csv', 'w')

# Header row
file.write('NAME, AGE, SPORTS')
file.write('\n')

# Rows data
for olympian in olympians:
    rows = f'{olympian[0]}, {olympian[1]}, {olympian[2]}'

    file.write(rows)
    file.write('\n')

file.close()
```

#### By Csv Module

```python
import csv 

Employees = [('Ismail', 23, 'Data Sceitist'),
             ('Ahmed', 30, 'Civil Engineer'),
             ('Abdalle', 22, 'Accountant'),
             ('Sakeriye', 24, 'HR'),
             ('Fowsiyo', 22, 'Investment'),
             ('Osman', 50, 'Head of Logistic'),
             ('Abdinasir', 42, 'Logistic'),
             ('Abdisammad', 66, 'Security'),
             ]


file = open('Dahab_Employees.csv', 'w')

writer = csv.writer(file)

# Writerow expects a list, not comma seperated string, And no need to write '\n'

writer.writerow(['Name', 'Age', 'Role'])

for employee in Employees:
    # Especially easy to write a row if we already have it as a list value
    writer.writerow(employee)

file.close()
```
