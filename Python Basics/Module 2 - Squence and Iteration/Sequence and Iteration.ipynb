```python
names = ['Ismail', 'Ahmed', 'Ismail', 'Mohamed', 'Ahmed']
dict = {}

# accum = 0

for name in names:
    if name not in dict:
        dict[name] = 1
    else:
        dict[name] = dict[name] + 1
print(dict)
```

### Acumulator Pattern wuth conditionals and Accumulating a maximum value.

```python
# Calculating number of characters Excepting whitespace
phrase = 'What a wonderful day to program'
total = 0

for char in phrase:
    if char != ' ':
        total = total + 1
print(total)
```

```python
# Calculating number of vowels
s = 'What if we want to the zoo'

x = 0

for vowels in s:
    if vowels in ['a', 'e', 'i', 'o', 'u']:
        x = x + 1

print(f'The number of vowels inside the sequance is: {x}')
```

```python
# Example 3: Calculating the max value
numbers = [9, 10, 22, 30,8, 10, 0]

max_num = numbers[0]

for number in numbers:
    if number > max_num:
        max_num = number

print(f'The maximum number of this sequance is: {max_num}')
```

### Accumulating Lists and Strings

```python
# Example 1: Accumalator Pattern With Lists
numbers = [10, 20, 30]

accum = []

for num in numbers:
    x = num ** 2
    accum.append(x)

print(accum)
```

```python
# Example 2: Accumulator Pattern With String

greeting = '  Hello, world  '

total_char = 0

for i in greeting:
    if i != ' ':
        total_char += 1
print(total_char)
```

```python
# Example 3
greeting = 'Helloo'

accum = ''

for char in greeting:
    accum = accum + char + '-'

print(accum)
```
